```python
# convert units based on user choice
def convert_units():
    """Convert length, weight, or volume units."""
    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter your choice (1-3): ")
        value = float(input("Enter the value to convert: "))

        if choice == "1":
            print("Choose length conversion:")
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            option = input("Enter your choice (1-2): ")

            if option == "1":
                result = value * 3.28084
                print("Converted value:", result, "ft")
            elif option == "2":
                result = value / 3.28084
                print("Converted value:", result, "m")
            else:
                print("Invalid length option")

        elif choice == "2":
            print("Choose weight conversion:")
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            option = input("Enter your choice (1-2): ")

            if option == "1":
                result = value * 2.20462
                print("Converted value:", result, "lbs")
            elif option == "2":
                result = value / 2.20462
                print("Converted value:", result, "kg")
            else:
                print("Invalid weight option")

        elif choice == "3":
            print("Choose volume conversion:")
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            option = input("Enter your choice (1-2): ")

            if option == "1":
                result = value * 0.264172
                print("Converted value:", result, "gal")
            elif option == "2":
                result = value / 0.264172
                print("Converted value:", result, "L")
            else:
                print("Invalid volume option")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Please enter a numeric value only")


convert_units()
```

```python
# return sum of numbers
def find_sum(numbers):
    """Return the sum of the list."""
    return sum(numbers)

# return average of numbers
def find_average(numbers):
    """Return the average of the list."""
    return sum(numbers) / len(numbers)

# return largest number
def find_maximum(numbers):
    """Return the maximum value."""
    return max(numbers)

# return smallest number
def find_minimum(numbers):
    """Return the minimum value."""
    return min(numbers)


try:
    user_input = input("Enter numbers separated by spaces: ")
    numbers = list(map(float, user_input.split()))

    if len(numbers) == 0:
        print("The list is empty")
    else:
        print("Choose operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            print("Sum =", find_sum(numbers))
        elif choice == "2":
            print("Average =", find_average(numbers))
        elif choice == "3":
            print("Maximum =", find_maximum(numbers))
        elif choice == "4":
            print("Minimum =", find_minimum(numbers))
        else:
            print("Invalid operation")

except ValueError:
    print("Please enter valid numbers only")
```

 Exercise on List Manipulation

Extract Every Other Element

```python
# return every other element from the list
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

Slice a Sublist

```python
# return sublist from start to end inclusive
def get_sublist(lst, start, end):
    return lst[start:end+1]

print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

Reverse a List Using Slicing

```python
# reverse list using slicing
def reverse_list(lst):
    return lst[::-1]

print(reverse_list([1, 2, 3, 4, 5]))
```

Remove the First and Last Elements

```python
# remove first and last elements
def remove_first_last(lst):
    return lst[1:-1]

print(remove_first_last([1, 2, 3, 4, 5]))
```

Get the First n Elements

```python
# return first n elements
def get_first_n(lst, n):
    return lst[:n]

print(get_first_n([1, 2, 3, 4, 5], 3))
```

Extract Elements from the End

```python
# return last n elements
def get_last_n(lst, n):
    return lst[-n:]

print(get_last_n([1, 2, 3, 4, 5], 2))
```

Extract Elements in Reverse Order

```python
# return elements in reverse order by skipping one
def reverse_skip(lst):
    return lst[-2::-2]

print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

 Exercise on Nested List

Flatten a Nested List

```python
# flatten one-level nested list
def flatten(lst):
    result = []
    for sublist in lst:
        result.extend(sublist)
    return result

print(flatten([[1, 2], [3, 4], [5]]))
```

Accessing Nested List Elements

```python
# access an element using given indices
def access_nested_element(lst, indices):
    value = lst
    for index in indices:
        value = value[index]
    return value

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))
```

Sum of All Elements in a Nested List

```python
# find sum of all nested elements
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

Remove Specific Element from a Nested List

```python
# remove all occurrences of elem from nested list
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            cleaned = remove_element(item, elem)
            result.append(cleaned)
        else:
            if item != elem:
                result.append(item)
    return result

print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

Find the Maximum Element in a Nested List

```python
# find maximum element in nested list
def find_max(lst):
    max_value = float("-inf")
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)
    return max_value

print(find_max([[1, 2], [3, [4, 5]], 6]))
```

Count Occurrences of an Element in a Nested List

```python
# count occurrences of elem in nested list
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

Flatten a List of Lists of Lists

```python
# deeply flatten nested list
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

Nested List Average

```python
# find average of all nested elements
def average_nested(lst):
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)

print(average_nested([[1, 2], [3, 4], [5, 6]]))
```

Matrix Operations

```python
import numpy as np

# define matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. prove A.A^-1 = I
A_inv = np.linalg.inv(A)
identity_result = np.dot(A, A_inv)
print("1.\n", identity_result)

# 2. prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("2. AB =\n", AB)
print("2. BA =\n", BA)

# 3. prove (AB)^T = B^T A^T
left_side = (np.dot(A, B)).T
right_side = np.dot(B.T, A.T)
print("3. (AB)^T =\n", left_side)
print("3. B^T A^T =\n", right_side)

# represent system of equations
coeff = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

const = np.array([-1, -3, 9])

# solve using inverse method
solution = np.dot(np.linalg.inv(coeff), const)
print("Solution using inverse method =", solution)

# solve using np.linalg.inv directly
inv_coeff = np.linalg.inv(coeff)
solution2 = np.dot(inv_coeff, const)
print("Solution using np.linalg.inv =", solution2)
```

Experiment: How Fast is NumPy?

```python
import time
import numpy as np

# create large Python lists
list1 = list(range(1000000))
list2 = list(range(1000000))

# create large NumPy arrays
arr1 = np.array(list1)
arr2 = np.array(list2)

# 1. element-wise addition using Python lists
start = time.time()
list_add = [list1[i] + list2[i] for i in range(len(list1))]
end = time.time()
print("Python list addition time:", end - start, "seconds")

# 1. element-wise addition using NumPy
start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy addition time:", end - start, "seconds")

# 2. element-wise multiplication using Python lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(len(list1))]
end = time.time()
print("Python list multiplication time:", end - start, "seconds")

# 2. element-wise multiplication using NumPy
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy multiplication time:", end - start, "seconds")

# 3. dot product using Python lists
start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(len(list1)))
end = time.time()
print("Python list dot product time:", end - start, "seconds")

# 3. dot product using NumPy
start = time.time()
dot_arr = np.dot(arr1, arr2)
end = time.time()
print("NumPy dot product time:", end - start, "seconds")
```
